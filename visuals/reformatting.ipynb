```python
df = pd.read_csv('./gpttaxonomyupdated.csv')
df
```

```python
mask = df['Skill_ID'].isin({13, 71, 112})
df = df.loc[~mask]
```

```python
df = df.rename(columns = {'Original Skill Statement': 'requirement', 'New Sub-Skill Statement': 'skill'})
df['old_skill_id'] = df['Skill_ID']
```

```python
requirement_mapping = pd.read_csv('./DataMappingSkills.csv')
requirement_mapping
```

```python
df =pd.merge(df, requirement_mapping[['Skill_ID','requirement_id']], on='Skill_ID')
```

```python
df = df.rename(columns={'Skill_ID': 'skill_id'})
df['skill_id'] = df.index + 1
df = df.drop(columns={'Breakdown Sequence Number'})
```

```python
cols = df.columns.tolist()
cols = [cols[1]] + [cols[0]] + cols[2:]
cols = cols[-1:] + cols[:-1]
df = df[cols]
```

```python
df.loc[df['skill'].str.contains(' CO2'), 'skill'] = df['skill'].str.replace(' CO2', ' carbon dioxide (CO2)', regex=False)
```

```python
df.loc[df['skill'].str.contains(' CCS '), 'skill'] = df['skill'].str.replace(' CCS ', ' carbon capture and storage ')
```

```python
df.loc[df['skill'].str.contains('CCS-related'), 'skill'] = df['skill'].str.replace('CCS-related', 'carbon capture and storage related')
```

```python
df.loc[df['skill'].str.contains('ATR'), 'skill']
```

```python
df = df.drop(columns={'Assigned Taxonomy Location','Additional Notes/Comments','Highest Level Taxonony Cluster','old_skill_id'})
```

```python
df.to_csv('./mnt/data/reformatted_skills.csv')
```
